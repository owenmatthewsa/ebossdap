```python
ext = dext()
OII_corr = ext.extinguish(1 /(3728 * 0.0001 * (1+ df_comp['z'])),Ebv=df_comp['EBV'])
Hb_corr = ext.extinguish(1 / (4863 * 0.0001 * (1+ df_comp['z'])), Ebv=df_comp['EBV'])
OIII_corr = ext.extinguish(1 / (5008 * 0.0001 * (1+ df_comp['z'])), Ebv=df_comp['EBV'])
Ha_corr = ext.extinguish( 1/ (6565 * 0.0001 * (1+ df_comp['z'])), Ebv=df_comp['EBV'])
NII_corr =  ext.extinguish(1 / (6584 * 0.0001 * (1+ df_comp['z'])), Ebv=df_comp['EBV'])
SII_corr = ext.extinguish(1 / (6724 * 0.0001 * (1+ df_comp['z'])), Ebv=df_comp['EBV'])

'''
OII_A_lambda = (extinction_curve.evaluate(3728 * 0.0001 * (1+ df_comp['z'])) ) * df_comp['EBV']
OII_flux_correction = 10**(OII_A_lambda / (-2.5))

Hb_A_lambda = (extinction_curve.evaluate(4863 * 0.0001 * (1+ df_comp['z'])) ) * df_comp['EBV']
Hb_flux_correction = 10**(Hb_A_lambda / (-2.5))

OIII_A_lambda = (extinction_curve.evaluate(5008 * 0.0001 * (1+ df_comp['z'])) ) * df_comp['EBV']
OIII_flux_correction = 10**(OIII_A_lambda / (-2.5))

Ha_A_lambda = (extinction_curve.evaluate(6565 * 0.0001 * (1+ df_comp['z'])) ) * df_comp['EBV']
Ha_flux_correction = 10**(Ha_A_lambda / (-2.5))
'''

df_comp['OII_sdss'] = df_comp['OII_sdss'] / (OII_corr)
df_comp['Hb_sdss'] = df_comp['Hb_sdss'] / (Hb_corr)
df_comp['OIII_sdss'] = df_comp['OIII_sdss'] / (OIII_corr)
df_comp['Ha_sdss'] = df_comp['Ha_sdss'] / (Ha_corr)
df_comp['NII_sdss'] = df_comp['NII_sdss'] / (NII_corr)
df_comp['SII_sdss'] = df_comp['SII_sdss'] / (SII_corr)
```

```python
df_comp['OII'] = df_comp['OII_3727'] + df_comp['OII_3729']
df_comp['OII_err'] = np.sqrt(df_comp['OII_3727_err']**2 + df_comp['OII_3729_err']**2)
df_comp['OIII_ew'] = df_comp['OIII_5008_ew']

df_comp['SII'] = df_comp['SII_6718'] + df_comp['SII_6732']
df_comp['SII_err'] = np.sqrt(df_comp['SII_6718_err']**2 + df_comp['SII_6732_err']**2)
#df_comp['NII_ew'] = df_comp['NII_6585_ew']


df_comp['OII_flux1'] = df_comp['OII']
df_comp['Hb_flux1'] = df_comp['Hb']
df_comp['OIII_flux1'] = df_comp['OIII_5008']
df_comp['Ha_flux1'] = df_comp['Ha']
df_comp['NII_flux1'] = df_comp['NII_6585']
df_comp['SII_flux1'] = df_comp['SII']


df_comp['OII_flux1_err'] = df_comp['OII_err']
df_comp['Hb_flux1_err'] = df_comp['Hb_err']
df_comp['OIII_flux1_err'] = df_comp['OIII_5008_err']
df_comp['Ha_flux1_err'] = df_comp['Ha_err']
df_comp['NII_flux1_err'] = df_comp['NII_6585_err']
df_comp['SII_flux1_err'] = df_comp['SII_err']
```

```python
ew_cols = [col.name for col in data.columns if col.name.endswith('_EW') and not col.name.endswith('_INT_EW')]

# Loop through each pair
for ew_col in ew_cols:
    int_ew_col = ew_col.replace('_EW', '_INT_EW')
    if int_ew_col in data.columns.names:
        ew_vals = data[ew_col]
        int_ew_vals = data[int_ew_col]

        # Filter out NaNs and invalid EW values
        mask = (~np.isnan(ew_vals)) & (~np.isnan(int_ew_vals)) & (ew_vals != -999.0) & (ew_vals != 0.0) 
        filtered_diff = ew_vals[mask] - int_ew_vals[mask]
        filtered_diff = filtered_diff[np.abs(filtered_diff) <= 50]

        if len(filtered_diff) > 0:
            plt.figure(figsize=(6, 4))
            plt.hist(filtered_diff, bins=100, density=True, alpha=1.0, color='C0')

            mu, std = norm.fit(filtered_diff)
            x = np.linspace(-50, 50, 500)
            p = norm.pdf(x, mu, std)
            plt.plot(x, p, '--', linewidth=2, label=f' = {mu:.2f},  = {std:.2f}',c='C1')

            plt.xlabel(f'Difference ({ew_col} - {int_ew_col})')
            plt.ylabel('Density')
            plt.title(f'{ew_col} - {int_ew_col} (50 clipped)')
            plt.legend()
            plt.tight_layout()
            plt.show()
```
