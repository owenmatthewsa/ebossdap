```python
#collumn creation for dataset 2 by Drake
plate = spall[1].data['PLATE'].byteswap().newbyteorder()
mjd = spall[1].data['MJD'].byteswap().newbyteorder()
fiber = spall[1].data['FIBERID'].byteswap().newbyteorder()
snrinit = spall[1].data['SN_MEDIAN'].byteswap().newbyteorder()
snr = snrinit[:,2]
snr_i = snrinit[:,3]

snr_all= spall[1].data['SN_MEDIAN_ALL'].byteswap().newbyteorder()
lameff = spall[1].data['LAMBDA_EFF'].byteswap().newbyteorder()

modelflux_g = spall[1].data['MODELFLUX'].byteswap().newbyteorder()[:,1]
modelflux_r = spall[1].data['MODELFLUX'].byteswap().newbyteorder()[:,2]
modelflux_i = spall[1].data['MODELFLUX'].byteswap().newbyteorder()[:,3]

#= -2.5 * np.log10(modelflux_r_err)

photo_color_gr = -2.5 * np.log10(modelflux_g/modelflux_r)
photo_color_ri = -2.5 * np.log10(modelflux_r/modelflux_i)

model_mag_r = 22.5 - 2.5 * np.log10(modelflux_r)



modelflux_g_err = np.power(spall[1].data['MODELFLUX_IVAR'].byteswap().newbyteorder()[:,1],-0.5)
modelflux_r_err = np.power(spall[1].data['MODELFLUX_IVAR'].byteswap().newbyteorder()[:,2],-0.5)
modelflux_i_err = np.power(spall[1].data['MODELFLUX_IVAR'].byteswap().newbyteorder()[:,3],-0.5)

mag_g_err = (2.5 / np.log(10)) * modelflux_g_err / modelflux_g
mag_r_err = (2.5 / np.log(10)) * modelflux_r_err / modelflux_r
mag_i_err = (2.5 / np.log(10)) * modelflux_i_err / modelflux_i
#mag_err = \frac{2.5}{\ln(10)} \times \frac{flux_err}{flux}
del modelflux_g
del modelflux_r
del modelflux_i

del modelflux_g_err
del modelflux_r_err
del modelflux_i_err

photo_color_gr_err = np.sqrt(mag_g_err**2 + mag_r_err**2)
photo_color_ri_err = np.sqrt(mag_r_err**2 + mag_i_err**2)


del mag_g_err
del mag_r_err
del mag_i_err

spectroflux_g = spall[1].data['SPECTROFLUX'].byteswap().newbyteorder()[:,1]
spectroflux_r = spall[1].data['SPECTROFLUX'].byteswap().newbyteorder()[:,2]
spectroflux_i = spall[1].data['SPECTROFLUX'].byteswap().newbyteorder()[:,3]
spectro_color_gr = -2.5 * np.log10(spectroflux_g/spectroflux_r)
spectro_color_ri = -2.5 * np.log10(spectroflux_r/spectroflux_i)


del spectroflux_g
del spectroflux_r
del spectroflux_i


f2mag_g =  spall[1].data['FIBER2MAG'].byteswap().newbyteorder()[:,1]
f2mag_r =  spall[1].data['FIBER2MAG'].byteswap().newbyteorder()[:,2]
f2mag_i =  spall[1].data['FIBER2MAG'].byteswap().newbyteorder()[:,3]

f2mag_g_err =  spall[1].data['FIBER2MAGERR'].byteswap().newbyteorder()[:,1]
f2mag_r_err =  spall[1].data['FIBER2MAGERR'].byteswap().newbyteorder()[:,2]
f2mag_i_err =  spall[1].data['FIBER2MAGERR'].byteswap().newbyteorder()[:,3]



f2mag_color_gr = f2mag_g - f2mag_r
f2mag_color_ri = f2mag_r - f2mag_i

del f2mag_g 
del f2mag_r 
del f2mag_i

f2mag_color_gr_err = np.sqrt(f2mag_g_err**2 + f2mag_r_err**2)
f2mag_color_ri_err = np.sqrt(f2mag_r_err**2 + f2mag_i_err**2)


del f2mag_g_err
del f2mag_r_err
del f2mag_i_err


#dataframe creation of Drake's data to get LOGM values which are not in my data
dfinit_a = pd.DataFrame({'plate':plate,'mjd':mjd,'fiber':fiber,'snr':snr,'lambda_eff':lameff, 'snr_all':snr_all,'snr_i':snr_i,
                         'photo_color_gr':photo_color_gr,'photo_color_gr_err':photo_color_gr_err,
                         'f2mag_color_gr':f2mag_color_gr,'f2mag_color_gr_err':f2mag_color_gr_err,
                         'spectro_color_gr':spectro_color_gr,
                         'photo_color_ri':photo_color_ri,'photo_color_ri_err':photo_color_ri_err,
                         'f2mag_color_ri':f2mag_color_ri,'f2mag_color_ri_err':f2mag_color_ri_err,
                         'spectro_color_ri':spectro_color_ri,'model_mag_r':model_mag_r})
dfgood = dfinit_a[dfinit_a['snr']>0].reset_index().drop(columns=['index'])



#collumn creation for dataset 2 by Drake
taginit = hdu2[1].data['PMF_String'].byteswap().newbyteorder()
tagtemp = [s.lstrip("0") for s in taginit]
tag = [s.replace('_','-') for s in tagtemp]
logm = hdu2[1].data['LOGMSTAR'].byteswap().newbyteorder()
ra = hdu2[1].data['RA'].byteswap().newbyteorder()
dec = hdu2[1].data['DEC'].byteswap().newbyteorder()
z = hdu2[1].data['Z'].byteswap().newbyteorder()
#dataframe creation of Drake's data to get LOGM values which are not in my data
dfinit = pd.DataFrame({'tag':tag,'RA':ra,'DEC':dec,'z':z})
dftemp = dfinit[dfinit['z']>0.001].reset_index().drop(columns=['index'])
```

```python
import itertools

# Unique Combination Pairs for list of elements
def uniqueCombinations(list_elements):
    l = list(itertools.combinations(list_elements, 2))
    s = set(l)
    # print('actual', len(l), l)
    return list(s)
```

```python
uniques1 = []
uniques2 = []
for i in np.arange(0,len(df_fin['tags']),1):
    #print(i)
    a,b = zip(*uniqueCombinations(df_fin['tags'][i]))
    uniques1.extend(list(a))
    uniques2.extend(list(b))
```

```python
import sys
def sizeof_fmt(num, suffix='B'):
    ''' by Fred Cirera,  https://stackoverflow.com/a/1094933/1870254, modified'''
    for unit in ['','Ki','Mi','Gi','Ti','Pi','Ei','Zi']:
        if abs(num) < 1024.0:
            return "%3.1f %s%s" % (num, unit, suffix)
        num /= 1024.0
    return "%.1f %s%s" % (num, 'Yi', suffix)

for name, size in sorted(((name, sys.getsizeof(value)) for name, value in list(
                          locals().items())), key= lambda x: -x[1])[:10]:
    print("{:>30}: {:>8}".format(name, sizeof_fmt(size)))
```

```python
lines = ['Hb', 'NeV','OII','NII_6585','OIII_5008','Ha','OIII_4364', 'HeII_4687',
         'Hb_ew', 'NeV_ew','OII_ew','NII_6549_ew','OIII_5008_ew','Ha_ew','OIII_4364_ew', 'HeII_4687_ew']

#8 is where Hb_ew is 

line = lines[1]

savedf = pd.read_csv('saved_matched_fluxes.csv')
savedf['diff'] =  (np.divide(savedf[line+'_x'],savedf[line+'_y']))
boolarray = np.random.choice(2,len(savedf['diff']))
savedf['diff_rand'] = np.where(boolarray==1,(np.divide(savedf[line+'_x'],savedf[line+'_y'])),(np.divide(savedf[line+'_y'],savedf[line+'_x'])))
```

```python
#print(savedf.columns)
savedf['S/N1'] = (np.divide(savedf[line+'_x'],savedf[line+'_err_x']))
savedf['S/N2'] = (np.divide(savedf[line+'_y'],savedf[line+'_err_y']))
savedf['flux1'] = savedf[line+'_x']
savedf['flux2'] = savedf[line+'_y']
savedf['zspeed1'] = savedf['Z_x'] * 299792
savedf['zspeed2'] = savedf['Z_y'] * 299792
```

```python
df1 = savedf[np.abs(savedf['diff']) <= 50]
df2 = df1[np.abs(df1['S/N1']) <= 50]
df3 = df2[df2['flux1']!=0.0]
df4 = df3[df3['flux2']!=0.0]
df5 = df4[df4['flux2']!=-999.0]
df6 = df5[df5['flux2']!=-999.0]
```

```python
df7 = df6[['diff','diff_rand','flux1','flux2','S/N1','S/N2', 'zspeed1','zspeed2']]
```

```python
x = np.linspace(0,15,500)
y = np.sqrt(2)/(x)
x2 = np.linspace(0,15,150)
y2 = np.sqrt(2)/(x)
```

```python
snrs = []
std_top = []
std_bot = []
means = []
std = []
for i,ix in enumerate(x2[1:]):
    #print(x[i], ix)
    dfless = df7[df7['S/N1']<ix]
    dfslice = dfless[dfless['S/N1']> x2[i]]
    snrs.append((ix+x2[i])/2)
    std.append(np.std(np.log(dfslice['diff_rand'])))
    if len(dfslice) != 0:
        std_top.append(np.percentile(np.log(dfslice['diff_rand']), 68.27+15.7))
        std_bot.append(np.percentile(np.log(dfslice['diff_rand']), 15.7))
    else:
        std_top.append(np.nan)
        std_bot.append(np.nan)
    means.append(np.mean(np.log(dfslice['diff_rand'])))
```
