```python
# Imports
import sys
import os
import numpy as np
import pandas as pd
from astropy.io import fits
import astropy.coordinates as acoords
from astropy.io import fits
import glob
from matplotlib import pyplot as plt
np.set_printoptions(suppress=True)
import time
from IPython import embed
import astropy.units as u
import astropy.constants
import seaborn as sns
import matplotlib.lines as mlines
import matplotlib.patheffects as pe
def plot_prettier(dpi=300, fontsize=9, figsize = (11,8.5)):
    # import pyplot and set some parameters to make plots prettier
    import matplotlib.pyplot as plt
    plt.rc("savefig", dpi=dpi)       # dpi resolution of saved image files
    plt.rc('font', size=fontsize)    # fontsize
    plt.rc('xtick', direction='in')  # make axes ticks point inward
    plt.rc('ytick', direction='in')
    plt.rc('xtick.major', pad=10) 
    plt.rc('xtick.minor', pad=5)
    plt.rc('ytick.major', pad=10) 
    plt.rc('ytick.minor', pad=5)
    plt.rc('lines', dotted_pattern = [0.5, 1.1]) # fix dotted lines
    plt.rc('figure', figsize=figsize,facecolor='white')
    
    return
plot_prettier(150, fontsize=18)

np.set_printoptions(suppress=True)
#quick cell stopper for debugging
class StopExecution(Exception):
    def _render_traceback_(self):
        return []
#fixing warnings to make my life easier
import warnings

def fxn():
    warnings.warn("deprecated", DeprecationWarning)

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    fxn()
```

```python
#creating the lines to be drawn for BPT
bptline1x = np.linspace(-2.0,-0.1,100)
bptline1y = 0.438/(bptline1x + 0.023) + 1.222
bptline2y = np.linspace(-0.65,0.9,100)
bptline2x = -0.39 * bptline2y**4 - 0.582*bptline2y**3 - 0.637*bptline2y**2-0.048*bptline2y-0.119
bptline3x = np.linspace(-0.24,0.5,100)
bptline3y = 0.95*bptline3x +0.56


#creating the lines to be drawn for MEx
mexline1x = np.linspace(7,10,100)#+ 0.2*(1+zint)
mexline1y = (0.375/(mexline1x-10.5)) + 1.14
mexline2x = np.linspace(10,13,100)#+ 0.2*(1+zint)
x = mexline2x #+ 0.2*(1+zint)
a0 = 410.24
a1 = -109.333
a2 = 9.71731
a3 = -0.288244
mexline2y = a0 + a1*x + a2*x**2 + a3*x**3
mexline3x = np.linspace(9.59,13,100)#+ 0.2*(1+zint)
x = mexline3x
a0 = 352.066
a1 = -93.8249
a2 = 8.32651
a3 = -0.246416
mexline3y = a0 + a1*x + a2*x**2 + a3*x**3


#creating the lines to be drawn for Blue
bluline1x = np.linspace(-2.0,0.9,100)
bluline1y = (0.11/(bluline1x-0.92)) + 0.85
bluline2x = np.linspace(0.09,1.06,100)
bluline2y = -(bluline2x-1.0)**2 - 0.1*bluline2x + 0.25
bluline3x = np.linspace(0.09,1.06,100)
bluline3y = (bluline2x-0.2)**2 - 0.6
bluline4x = np.linspace(-2.0,0.72,100)
bluline4y = np.ones_like(bluline4x) * 0.3
bluline5x = np.linspace(0.72,2,100)
bluline5y = 0.95*(bluline5x) - 0.4



#creating the lines to be drawn for TrEW

trline1x = np.linspace(0.5,3.0,100)
trline1y = -3.25 +  3.20*trline1x   - 0.62*(trline1x**2)
trline2x = np.linspace(0.5,1.641,100)
trline2y = np.ones_like(trline2x) * 0.332
```
