```python
oh_theoretical_all = np.linspace(7.6, 9.0, 10_000)
oh_sol = 8.69
x = oh_theoretical_all - oh_sol
r23_theoretical_all = np.poly1d(np.array([0.527, -1.569, -1.652, -0.421])[::-1])(x)
max_r23 = np.nanmax(r23_theoretical_all[~np.isinf(r23_theoretical_all)])
min_r23 = np.nanmin(r23_theoretical_all[~np.isinf(r23_theoretical_all)])
r23_lowb_min = r23_theoretical_all[0]

r23_turn = 8.07 # 12+log(O/H) value at turnover of r23 diagnostic

def get_branch(o32, r23_turn=r23_turn): # o32 diagnostic is monotonic, so choose branch based on that
    
    oh_sol = 8.69
    x = r23_turn - oh_sol
    o32_split = np.poly1d(np.array([-0.691, -2.944, -1.308])[::-1])(x)
    
    if o32 > o32_split: return 'lower'
    elif o32 <= o32_split: return 'upper'
    else: return 'error'
    

def get_r23_metal(r23, branch, r23_turn=r23_turn):
    
    """returns 12+log(O/H) from Curti et al. 2017 calibrations"""

    if branch=='lower' and r23 >= r23_lowb_min:
        oh_theoretical = np.linspace(7.6, r23_turn, 10_000)
    elif branch=='upper':
        oh_theoretical = np.linspace(r23_turn, 9.0, 10_000)
    elif branch=='lower' and r23 < r23_lowb_min:
        return np.nan
    elif branch=='error':
        return np.nan

    oh_sol = 8.69
    x = oh_theoretical - oh_sol
    r23_theoretical = np.poly1d(np.array([0.527, -1.569, -1.652, -0.421])[::-1])(x)

    metal = oh_theoretical[np.argmin(abs(r23_theoretical - r23))]

    return metal

def bpt_kewley01(logniiha):
    '''
    Defining the BPT AGN/SF dividing line from Kewley 2001.
    '''    
    return 0.61/(logniiha - 0.47) + 1.19    

def mass_excitation_j14(logmass):
    '''
    Return the Mass Excitation AGN/SF dividing line from Juneau et al. 2014
    '''
    y_upper = np.zeros(len(logmass))
    for i,m in enumerate(logmass):
        if m <= 10:
            y_upper[i] = 0.375/(m - 10.5) + 1.14
        else:
            y_upper[i] = 410.24 + -109.333*m + 9.71731*m**2 + -0.288244*m**3
            
    y_lower = np.zeros(len(logmass))
    for i,m in enumerate(logmass):
        if m <= 9.6:
            y_lower[i] = 0.375/(m - 10.5) + 1.14
        else:
            y_lower[i] = 352.066 + -93.8249*m + 8.32651*m**2 + -0.246416*m**3
    
    return y_upper, y_lower
```

```python

#df_samp = df.sample(len(df)) # option to run on fewer than all objects
min_flux, max_flux = 0, 10000
lines = ['OII', 'OIII', 'Hbeta', 'Halpha', 'NII']
for line in lines:
    df = df[(df[line] > min_flux) # choose positive fluxes
                      & (df[line] < max_flux) # clip crazy high fluxes
                      & (df[line] / df[f'{line}_err'] > 3)] # clip low S/N objects

df['R3'] = np.log10(df.OIII / df.Hbeta)
df['N2'] = np.log10(df.NII / df.Halpha)



cut1 = df[np.where((-0.39 * df['R3']**4 - 0.582*df['R3']**3 - 0.637*df['R3']**2-0.048*df['R3']-0.119) >df['N2'] ,True,False)]
cut1_1 = cut1[np.where((0.438/(cut1['N2'] + 0.023) + 1.222)<cut1['R3'],True,False)]
BPT_composites = cut1_1[cut1_1['N2'] >-1.4]
cut2 = df[np.where((0.438/(df['N2'] + 0.023) + 1.222)>df['R3'],True,False)]
BPT_SFGs = cut2[np.where(cut2['N2']<-0.1,True,False)]
cut3 = df[np.invert(df.index.isin(pd.concat([BPT_composites,BPT_SFGs]).index))]
BPT_LIERs = cut3[np.where((cut3['R3'] < (0.95*cut3['N2'] +0.56)),True,False)]
BPT_AGNs = cut3[np.where((cut3['R3'] > (0.95*cut3['N2'] +0.56)),True,False)]

'''#Maps BPT classification onto final dataframe
df.loc[df.tag.isin(BPT_AGNs.tag),'BPT_class'] = 0
df.loc[df.tag.isin(BPT_SFGs.tag),'BPT_class'] = 1
df.loc[df.tag.isin(BPT_composites.tag),'BPT_class'] = 2
df.loc[df.tag.isin(BPT_LIERs.tag),'BPT_class'] = 3'''

df_samp_init = BPT_SFGs #df_samp[(r3 < bpt_kewley01(n2)) * (r3 < mass_excitation_j14(df_samp.LOGM)[0])] # remove AGN
df_samp_init1 = df_samp_init[df_samp_init['redshift']< 0.4963]
df_samp = df_samp_init1[df_samp_init1['redshift']> 0.0005]

ids = df_samp.PMF_String
oii = df_samp.OII
oiii = df_samp.OIII
hb = df_samp.Hbeta
df_samp['metal'] = np.nan

df_samp['r23'] = np.log10((oii + oiii) / hb)
df_samp['o32'] = np.log10(oiii / oii)
df_samp = df_samp[(df_samp.r23 < max_r23) * (df_samp.r23 > min_r23)]

for ii in tqdm(range(len(df_samp))):
    idval = df_samp['PMF_String'].values[ii]
    r23_val = df_samp.r23.values[ii]
    o32_val = df_samp.o32.values[ii]
    branch = get_branch(o32_val)
    metal = get_r23_metal(r23_val, branch)
    df_samp.loc[df_samp['PMF_String']==idval, 'metal'] = metal
```
