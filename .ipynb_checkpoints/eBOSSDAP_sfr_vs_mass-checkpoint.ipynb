```python
blrgs_init = trimdf[ trimdf['code'].str.contains('BLRG')] # luminous red galaxy
elrgs_init = trimdf[ trimdf['code'].str.contains('ELRG')] # luminous red galaxy
llrgs_init = trimdf[ trimdf['code'].str.contains('LLRG')] # luminous red galaxy
elgs_init = trimdf[ trimdf['code'].str.contains('ELG')] #emission line galaxy
Tdss_init = trimdf[ trimdf['code'].str.contains('AGN')] #Time Domain Spectroscopic Survey (TDSS) & the SPectroscopic IDEntification of eROSITA Sources (SPIDERS) survey
qso_init = trimdf[ trimdf['code'].str.contains('QSO')] #emission line galaxy

blrgs= blrgs_init[(blrgs_init['LOGM']> 8) & (blrgs_init['LOGM']< 12) & (blrgs_init['LOGSFR']> -4) & (blrgs_init['LOGSFR']< 3)]
elrgs= elrgs_init[(elrgs_init['LOGM']> 8) & (elrgs_init['LOGM']< 12) & (elrgs_init['LOGSFR']> -4) & (elrgs_init['LOGSFR']< 3)]
llrgs= llrgs_init[(llrgs_init['LOGM']> 8) & (llrgs_init['LOGM']< 12) & (llrgs_init['LOGSFR']> -4) & (llrgs_init['LOGSFR']< 3)]
elgs= elgs_init[(elgs_init['LOGM']> 8) & (elgs_init['LOGM']< 12) & (elgs_init['LOGSFR']> -4) & (elgs_init['LOGSFR']< 3)]
Tdss= Tdss_init[(Tdss_init['LOGM']> 8) & (Tdss_init['LOGM']< 12) & (Tdss_init['LOGSFR']> -4) & (Tdss_init['LOGSFR']< 3)]
qso= qso_init[(qso_init['LOGM']> 8) & (qso_init['LOGM']< 12) & (qso_init['LOGSFR']> -4) & (qso_init['LOGSFR']< 3)]

fig = plt.figure(figsize=(12, 8))
fig.tight_layout(pad=-1.0)
plt.scatter(elrgs['LOGM'],elrgs['LOGSFR'],alpha=0.2,color="magenta",label='eBOSS LRG',zorder=10,s=2)
plt.scatter(blrgs['LOGM'],blrgs['LOGSFR'],alpha=0.2,color="xkcd:dark red",label='BOSS LRG',zorder=0,s=2)
plt.scatter(llrgs['LOGM'],llrgs['LOGSFR'],alpha=0.2,color="xkcd:burnt orange",label='BOSS LOWZ LRG',zorder=5,s=2)
plt.scatter(elgs['LOGM'],elgs['LOGSFR'],alpha=0.2,color="green",label='ELG',zorder=15,s=2)
plt.scatter(Tdss['LOGM'],Tdss['LOGSFR'],alpha=0.2,color='xkcd:dark yellow',label='TDSS/SPIDERS',zorder=20,s=2)
plt.scatter(qso['LOGM'],qso['LOGSFR'],alpha=0.2,color="blue",label='QSO',zorder=25,s=1)

for t in ages:
    x = np.log10(Mstars)
    y = a0 + a1*t - np.log10(1+(Mstars/10**(a2+a3*t))**(-a4))
    plt.plot(x,y,label = 'Age '+str(t)+' Gyr')
leg = plt.legend(loc=4,fontsize=12)
plt.ylabel(r'Log(SFR) [M$_{\odot}$ / yr]')
plt.xlabel(r'Log(M$_{*}$) [M$_{\odot}$]')
plt.xlim(8,12)
plt.ylim(-3,3)
leg.set_zorder(100)
for lh in leg.legendHandles: 
    lh.set_alpha(1)
    #lh.set_size(8)
plt.savefig(r"C:\Users\owenm\Documents\catalog_paper\\"+'mass_vs_sfr_classified.pdf',bbox_inches='tight')
plt.show()
```
